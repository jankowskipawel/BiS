ROZKLAD POZYCJI ZAWODNIKOW

```
positions = statistics['Pos']
count_methods = positions.value_counts()
plot = count_methods.plot.pie(y="Position",figsize=(10,10),autopct='%1.1f%%',textprops={'fontsize': 14,'color':'black'})
plot.set_title("Distribution of positions",color='black',fontsize=14)
```

BADANIE NORMALNOSCI ROZKLADU

```
alpha = 0.001
for x in statistics:
    if statistics[x].dtypes == object:
        continue
    k2,p = stats.normaltest(statistics[x])
    print(x)
    print(p)
    if p < alpha:
        print("result can be rejected")
    else:
        print("result cannot be rejected")
    print("=====")
```

ROWNOSC SREDNICH

```
AST_test = stats.ttest_ind(statistics['AST'],statistics['STL'])
print("assists and steals per game test p = ",AST_test[1])

BLK_test = stats.ttest_ind(statistics['BLK'],statistics['STL'])
print("blocks and steals per game test p = ",BLK_test[1])

AST_test = stats.ttest_ind(statistics['AST'],statistics['BLK'])
print("assists and blocks test p = ",AST_test[1])
```

BADANIE ROWNOSCI WARIANCJI

```
AST_test = stats.levene(statistics['AST'],statistics['STL'])
print("assists and steals per game test p = ",AST_test[1])

BLK_test = stats.levene(statistics['BLK'],statistics['STL'])
print("blocks and steals per game test p = ",BLK_test[1])

AST_test = stats.levene(statistics['AST'],statistics['BLK'])
print("assists and blocks test p = ",AST_test[1])
```

TEST DLA ZMIENNYCH ZALEZNYCH

```
FG_attempts = stats.ttest_rel(statistics['FG'],statistics['3PA'])
print("field goals and 3 point field goals attempts test p = ",FG_attempts[1])

fouls = stats.ttest_rel(statistics['DRB'],statistics['PF'])
print("defensive rebounds and personal fouls test p = ",fouls[1])
```

TESTOWANIE HIPOTEZ

Srednio zawodnik strzela 3 gole

```
goals_mean = statistics['FG'].mean()
print("real goals mean: ",goals_mean)
print(stats.ttest_1samp(statistics['FG'],3))
```

Sredni wiek zawodnika wynosi 26 lat

```
age_mean = statistics['Age'].mean()
print("real age mean: ",age_mean)
print(stats.ttest_1samp(statistics['Age'],26))
```

REGRESJA LINIOWA

```
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```
won = []
for i in range(len(statistics)):
    if i % 2 == 0:
        won.append(0)
    else:
        won.append(1)
statistics['won'] = won
statistics
```

```
X = statistics.drop("won",axis=1)
X = X.iloc[:,-24:]
y = statistics['won']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
df_t = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df_t
```

```
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
